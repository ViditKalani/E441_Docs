```python
import pandas as pd
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
import matplotlib.pyplot as plt
```

# Creating lineplot

```python
titanic.plot(subplots=True,figsize=(15,12),sharex=False)
plt.show()
```

```python
titanic.age.plot(figsize=(12,8))
```

# Customize the lineplot

```python
titanic.age.plot(figsize=(12,8),fontsize=13,c='red',linestyle='-')
plt.title('Titanic-Ages',fontsize=15)
plt.legend(loc=3,fontsize=15)
plt.xlabel('Passenger Number',fontsize=13)
plt.ylabel('Age',fontsize=13)
plt.show()
```

```python
#limit the passengers
titanic.age.plot(figsize=(12,8),fontsize=13,c='r',linestyle='-',xlim=(200,400),ylim=(0,80))
plt.show()
```

```python
#for creating ticks
xticks=[x for x in range(0,901,50)]
xticks
```

```python
yticks=[y for y in range(0,81,5)]
yticks
```

```python
titanic.age.plot(figsize=(12,8),fontsize=13,c='r',linestyle='-',xlim=(0,900),ylim=(0,80),xticks=xticks,yticks=yticks,rot=90)
plt.show()
```
