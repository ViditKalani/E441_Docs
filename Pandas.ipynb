# Position based indexing and slicing

```python
#1st row
insurance_df.iloc[0]
```

```python
type(insurance_df.iloc[0])
```

```python
#2nd row
insurance_df.iloc[1]
```

```python
#last row
insurance_df.iloc[-1]
```

```python
#rows at index position 1,2,3

insurance_df.iloc[[1,2,3]]
```

```python
#top 5 rows
insurance_df.iloc[:5]
```

```python
#last 5 rows
insurance_df.iloc[-5:]
```

```python
#all rows
insurance_df.iloc[:]
```

# slicing of rows and columns with iloc

```python
insurance_df.iloc[0,7]
```

```python
#first row and out of that first 3 columns
insurance_df.iloc[0,:3]
```

```python
#first row and column at index 1,3,5
insurance_df.iloc[0,[1,3,5]]
```

```python
# select top 3 rows and columns corresponding to those rows at index 1,3,5
insurance_df.iloc[:3,[1,3,5]]
```
